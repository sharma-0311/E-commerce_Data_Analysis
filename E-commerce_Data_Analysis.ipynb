```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go 
import plotly.io as pio  
import plotly.colors as colors

pio.templates.default = "plotly_white" #  Default theme white for graphs plot
```

```python
dataset = pd.read_csv("Sample - Superstore.csv", encoding='latin-1') # encoding is use to read the speacial characters
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.info()
```

# Checking Data Formats and Correcting

```python
dataset['Order Date'] = pd.to_datetime(dataset['Order Date'])
dataset['Ship Date'] = pd.to_datetime(dataset['Ship Date'])
```

```python
dataset.info()
```

# Adding New Date-Based Columns

```python
# Extract the Month, Year and Week( 0 for Monday, 6 for Sunday) from Dataset
dataset['Order Month'] = dataset['Order Date'].dt.month 
dataset['Order Year'] = dataset['Order Date'].dt.year
dataset['Order Day of Week'] = dataset['Order Date'].dt.dayofweek
```

```python
dataset.head()
```

# Monthly Sales Analysis

```python
sales_by_month = dataset.groupby('Order Month')['Sales'].sum().reset_index()  # Extracting Total Sales of Every Month

fig = px.line(sales_by_month, 
              x='Order Month', 
              y='Sales', 
              title='Monthly Sales Analysis')  # monthly sales trend shows by Line Chart
fig.show()
```

# Sales Analysis by Category

```python
sales_by_category = dataset.groupby('Category')['Sales'].sum().reset_index()\

fig = px.pie(sales_by_category, 
             values='Sales', 
             names='Category', 
             hole=0.5, 
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font=dict(size=24))

fig.show()
```

# Sales Analysis by Sub-Category

```python
sales_by_subcategory = dataset.groupby('Sub-Category')['Sales'].sum().reset_index()

fig = px.bar(sales_by_subcategory, 
             x='Sub-Category', 
             y='Sales', 
             title='Sales Analysis by Sub-Category')
fig.show()
```

# Monthly Profit Analysis

```python
profit_by_month = dataset.groupby('Order Month')['Profit'].sum().reset_index()

fig = px.line(profit_by_month, 
              x='Order Month', 
              y='Profit', 
              title='Monthly Profit Analysis')
fig.show()
```

# Profit Analysis by Category

```python
profit_by_category = dataset.groupby('Category')['Profit'].sum().reset_index()

fig = px.pie(profit_by_category, 
             values='Profit', 
             names='Category', 
             hole=0.5, 
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Profit Analysis by Category', title_font=dict(size=24))

fig.show()
```

# Profit Analysis by Sub-Category

```python
profit_by_subcategory = dataset.groupby('Sub-Category')['Profit'].sum().reset_index()

fig = px.bar(profit_by_subcategory, x='Sub-Category', 
             y='Profit', 
             title='Profit Analysis by Sub-Category')
fig.show()
```

# Sales and Profit Analysis by Customer Segment

```python
sales_profit_by_segment = dataset.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment
```

```python
color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], 
                     y=sales_profit_by_segment['Sales'], 
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], 
                     y=sales_profit_by_segment['Profit'], 
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment', yaxis_title='Amount')

fig.show()
```

# Analyse sales-to-profit ratio

```python
sales_profit_by_segment = dataset.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Ration of Sales to Profit
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
